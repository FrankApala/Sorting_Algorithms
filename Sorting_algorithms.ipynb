## 1) Merge Sort

## Working Principle 

Merge sort is a sorting algorithm that follows the divide-and-conquer paradigm. It works by recursively dividing the input array into two halves, sorting each half separately, and then merging the sorted halves into a single sorted array.

Here's how the merge sort algorithm works:

1)Divide the input array into two halves, recursively, until each sub-array contains only one element.

2)Merge the two sorted sub-arrays into a single sorted array by comparing the first element of each sub-array and placing the smaller one into the new array.

Repeat step 2 until all the elements of both sub-arrays have been merged into the new array.

The merge step of the algorithm can be done in different ways, but the most common approach is to use an auxiliary array to hold the merged elements, and then copy the merged array back to the original array. Below is a diagram illustrating the merge sort algorithm



The time complexity of merge sort is O(n*log(n)), which makes it one of the most efficient sorting algorithms, especially for large datasets. However, it has a higher space complexity compared to other sorting algorithms, as it requires additional memory to store the auxiliary array during the merging step.

Overall, merge sort is a reliable and efficient sorting algorithm that can be used in various applications that require sorting large datasets.

# Python Code for Merge Sort

The merge_sort() function is responsible for recursively dividing the input array into two halves and sorting each half using merge_sort(). The sorted halves are then merged together using the merge() function.

````python
def merge_sort(arr):
    if len(arr) <= 1:
        return arr

    # Divide the array into two halves
    mid = len(arr) // 2
    left = arr[:mid]
    right = arr[mid:]

    # Recursively sort each half
    left = merge_sort(left)
    right = merge_sort(right)

    # Merge the sorted halves into a single sorted array
    return merge(left, right)
````

The merge() function takes two sorted arrays as input and merges them together into a single sorted array. It uses two pointers to keep track of the current element in each input array, and compares them to determine the next element to add to the result array.

````python
def merge(left, right):
    merged = []
    i, j = 0, 0  # indices for left and right arrays

    # Iterate through both arrays and add the smaller element to the merged array
    for k in range(len(left) + len(right)):
        if i >= len(left):  # left array is exhausted
            merged.append(right[j])
            j += 1
        elif j >= len(right):  # right array is exhausted
            merged.append(left[i])
            i += 1
        elif left[i] < right[j]:
            merged.append(left[i])
            i += 1
        else:
            merged.append(right[j])
            j += 1

    return merged
````

# 2) Selection Sorting Algorithm

## Working Principle

This algorithm is the most naive sorting algorithm, his working principle is very simple.

In an unsorted list of n numbers, the algorithm will :

- Find the minimum of the list

- Switch the minimum with the first element of the list

- Do the same with the (n - 1) following terms until the end of the list




## Python Code

````python
def selSort(A):
    for i in range(len(A)): # i is the index below which the list is sorted
        mini = i
        for j in range(i,len(A)): # j is used to find the minimum of the unsorted part of the list
            if(A[j] < A[mini]):
                mini = j
        if(mini != i): # swap the i-th element of the list with the minimum that we found
            t = A[i]
            A[i] = A[mini]
            A[mini] = t
    return A
````

````python
A = [5,1,9,3,-4,7,-1,8,1,5,2,5,2,8,2,3,2,5,0,5,0,5,-10]

print(selSort(A))
````

## Proof of the algorithm

#### Our algorithm includes a loop invariant that will allow us to prove the correctness of the algorithm :

Let $i\in[\![ 0,n [\![$ (with $n$ the size of the list) be the step of the algorithm at which the i-th element of the list is swapped.

Let $A$ be a list of $n$ real numbers which will be denoted $A_k$ ($k\in[\![ 0,n [\![$)

Let $I_i = [\![0,i]\!]$, $I_j = ]\!]j,n[\![$ and $I_n = [\![ 0,n [\![$

##### Loop Invariant :

At the end of each step $i$ of the algorithm :
$$\forall j \in [\![0,i]\!], \forall k \in ]\!]j,n[\![, A_j \leq A_k \tag{E}$$

##### Proof :

We will show this by induction

**Base Case :** $i = 0$

At the end of the 0-th step, $A_0$ was swapped with the minimum of the list, so $\forall k \in I_n^*, A_0 \leq A_k$. The $P_0$ is true so we can make our induction hypothesis.

**Induction Hypothesis :**
$$\exists i \in I_n, \forall j \in I_i, \forall k \in I_j, A_j \leq A_k \tag{H}$$

At the ($i+1$)-th step of the algorithm the minimum of the sub-list $\{A_k \vert k \in [\![i+1,n[\![\}$ is swapped with $A_{i+1}$

So : 
$$
\begin{equation}
\begin{split}
& \forall k \in ]\!]i+1,n[\![, A_{i+1} \leq A_k \\
\implies & \forall j \in [\![0,i+1]\!], \forall k \in I_j, A_j \leq A_k \text{ , from } (\mathrm{H})\\
\implies & P_{i+1}
\end{split}
\end{equation}
$$

**Conclusion:** By mathematical induction, since the fact that $P_0$ is true and so is $(P_i \implies P_{i+1})$ is true, $P_i$ is true for all i

So we can now deduce from $(\mathrm{E})$ that at the end of the $(n-1)$-th step, all the real numbers of $A$ will be sorted.

# 3) Bubble Sort
Bubble sort is a sorting algorithm that compares two adjacent elements and swaps them until they are in the intended order.
Just like the movement of air bubbles in the water that rise up to the surface, each element of the array move to the end in each iteration. Therefore, it is called a bubble sort.




## Working Principle

Suppose we are trying to sort the elements in ascending order.
1. First Iteration (Compare and Swap)
    Starting from the first index, compare the first and the second elements.
    If the first element is greater than the second element, they are swapped.
    Now, compare the second and the third elements. Swap them if they are not in order.
    The above process goes on until the last element. 
2. Remaining Iteration
    The same process goes on for the remaining iterations.
    After each iteration, the largest element among the unsorted elements is placed at the end.
In each iteration, the comparison takes place up to the last unsorted element.
The array is sorted when all the unsorted elements are placed at their correct positions.



## Code

````python
# Bubble sort in Python

def bubbleSort(array):
      # loop to access each array element
    for i in range(len(array)):
        # loop to compare array elements
        for j in range(0, len(array) - i - 1):
          # compare two adjacent elements
          # change > to < to sort in descending order
            if array[j] > array[j + 1]:
                # swapping elements if elements
                # are not in the intended order
                temp = array[j]
                array[j] = array[j+1]
                array[j+1] = temp
    return array
````

````python
arr = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5]
sorted_arr = bubbleSort(arr)
print(sorted_arr)
````

#  4) Insertion Sort 

Insertion sort is a simple sorting algorithm that builds the final sorted array (or list) one item at a time by comparisons. It is much less efficient on large lists than more advanced algorithms such as quicksort, heapsort, or merge sort. However, this algorithm is pretty efficient on small set of data

## Working Principle

1.Initially, the first element in the array is considered to be the sorted subarray.

2.Iterate through the remaining unsorted elements one by one, and for each element:

3.Compare it with each element in the sorted subarray, starting from the rightmost element.

4.If the element is smaller than the current element in the sorted subarray, shift that element one position to the right to make space for the new element.

5.Repeat step 4 until you find the correct position for the new element.

6.Insert the new element into its correct position in the sorted subarray.

7.Repeat steps 2-6 for all remaining unsorted elements.

8.Once all elements have been inserted into the sorted subarray, the array is sorted.


## Algorithm in action

As you can see, the sorted elements are in black, and the current element is in red. As we cross the array, a comparison is done between the current element(in red) and the sorted elements. if the current element is less than one of the sorted elements, a swap is done.

Insertion-sort-example.gif

## Code

````python
def insertion_sort(arr):
    n = len(arr)  #We are storing the size of the array
    for i in range(1, n):
        unsorted_value = arr[i]  #We store the current element to do the sort
        j = i - 1  # We are also storing the index of the previous element
        while j >= 0 and arr[j] > unsorted_value:  #Which means why we have not visisted the sorted array:
            arr[j+1] = arr[j]  #We shift the values
            j -= 1 #We move one step back in the sorted list
        arr[j+1] = unsorted_value #We swap with the key
    return arr
````

````python
arr = [3, 1, 4,-2, 1, 5,-7, 9, 2, 6, 5, 3, 5]
sorted_arr = insertion_sort(arr)
print(sorted_arr)
````

# 5) Quicksort

## Working principle

Here are the general steps of the quicksort algorithm:
1.	Select a pivot element from the array. This can be done in many ways, but a common approach is to choose the first or last element in the array.
2.	Partition the array into two sub-arrays based on the pivot element. All elements less than the pivot are placed in the left sub-array, and all elements greater than or equal to the pivot are placed in the right sub-array.
3.	Recursively apply steps 1 and 2 to the left and right sub-arrays until the sub-arrays have length 0 or 1, at which point they are considered sorted.
4.	Concatenate the sorted left sub-array, the pivot element, and the sorted right sub-array to produce the final sorted array.

Below is an animation demonstating quicksort

## Code

````python
def partition(arr, first, last, pivot):
    arr[pivot], arr[last] = arr[last], arr[pivot]
    j = first
    for i in range(first, last):
        if arr[i] <= arr[last]:
            arr[i], arr[j] = arr[j], arr[i]
            j += 1
    arr[last], arr[j] = arr[j], arr[last]
    return j

def quicksort(arr, first, last):
    if first < last:
        pivot = first # choose a pivot element
        pivot = partition(arr, first, last, pivot) # partition the array around the pivot element
        quicksort(arr, first, pivot-1) # recursively sort the left subarray
        quicksort(arr, pivot+1, last) # recursively sort the right subarray
````

````python
arr = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5]
quicksort(arr,0,(len(arr)-1))
print(arr)
````

## Proof of the algorithm

#### Correctness of the partitioning algorithm :

Let $S$ be a subarray of the array that we want to sort and $S_i, i \in [\![0,n]\!]$ be the i-th term of $S$

Let $S_p$ be the pivot (wich is the same as $S_n$ here)

Let $i,j \in [\![0,n-1]\!]$ be two counter which will start at $0$ and be incremented during the execution of the algorithm

##### Loop Invariant :

At the end of each step of the algorithm :
$$ \forall k \in [\![0,i]\!], (k < j \implies S_k \leq S_p) \land (j \leq k \leq i \implies S_k > S_p) \tag{E}$$

##### Proof :

We will show this by induction

**Base Case :** $i = 0$

If $S_0 \leq S_p$ then $j$ is incremented (nothing is swapped because $i=j$)

So for $k = 0$ :
$$
\begin{equation}
\begin{split}
(j=0) & \implies (j \leq k \leq i) \land (S_0 > S_p) \\
\text{and } (j=1) & \implies (k<j) \land (S_0 \leq S_p)
\end{split}
\end{equation}
$$

So $P_0$ is true.

**Induction Hypothesis :**
$$\exists i \in [\![0,n-1]\!], \forall k \in [\![0,i]\!], (k < j \implies S_k \leq S_p) \land (j \leq k \leq i \implies S_k > S_p) \tag{H}$$

**Case $P_{i+1}$ :**

If $S_{i+1} \leq S_p$ then $S_{i+1}$ and $S_j$ are swapped and $j$ is incremented.

So after this :
$$S_{j-1} \leq S_p \tag{1}$$
And we know from $(\mathrm{H})$ that before the incrementation of $j$ : $j \leq j \leq i$ which implies $S_j > S_p$
So :
$$S_{i+1} > S_p\tag{2}$$

Otherwise if $S_{i+1} > S_p$ then nothing happens and $S_{j-1} \leq S_p$ from $(\mathrm{H})$

So $P_{i+1}$ is true (from the reunion of $(\mathrm{H})$, $(\mathrm{1})$ and $(\mathrm{2})$)


**Conclusion:** By mathematical induction, since the fact that $P_0$ is true and so is $(P_i \implies P_{i+1})$ is true, so $(\mathrm{E})$ is true.

So at the $(n-1)$-th step we know from $(\mathrm{E})$ that each term before the j-th term is lower or equal than $S_p$ and that $S_j$ and each term after it is greater than $S_p$.

So if we finally swap $S_j$ and $S_p$, we have successfully partitioned our subarray $S$.

##### Convergence :

The Quicksort Algorithm is a recursive algorithm and on each recursion step, the array that we want to sort is partitioned into 3 subarrays (left, right and the pivot) which are themselves partitioned into 3 subsubarrays at the next recursion step. Thus, the sizes of the sub-arrays decrease at each recursion step and will obviously converge to 1 since there is a condition on the size of the array to divide at the beginning of the algorithm.

Indeed the if statement (first < last) ensure that only arrays of size 2 or more can be subdivided.

##### Proof :

So we saw that the algorithm will divide the array to sort into subarrays of size 1 wich are then reassembled. Let's prove that these subarrays are sorted by induction:

Let $A$ be the array that we want to sort and $A(i), i \in [\![0,n]\!]$ be the i-th term of $A$

Let $S_i$ be the i-th subarray of the array that we want to sort and $S_i(j), j \in [\![0,len(S_i)]\!]$ be the j-th term of $S_i$

Let $S_i^j$ be the j-th subsubarrays of the i-th subarray of $A$ at a recursion step

**Loop Invariant :**

At each recursion step :

Let $m$ be the number of subarrays who were created.

$$\forall(i,j)\in [\![0,m[\![^2,(i<j)\implies (\forall (k_1,k_2),S_i(k_1) \leq S_j(k_2)) \tag{E}$$

**Base case :**

First recursion step :

$$
\begin{equation}
\begin{cases}
\forall k, S_0(k) < A(p) \\
S_1 = {A(p)} \\
\forall k, A(p) \leq S_2(k)
\end{cases}
\end{equation}
$$

So $P_0$ is true. We can make our induction hypothesis.

($S_0$ or $S_2$ could be the zero set but it doesn't change the statement)

**Induction Hypothesis :**

It exists some recursion step $r$ where the following statement is true :
$$\forall(i,j)\in [\![0,m[\![^2,(i<j)\implies (\forall (k_1,k_2),S_i(k_1) \leq S_j(k_2)) \tag{H}$$

**Case $P_{r+1}$ :**

Each subarray from the previous step is divided in 3 such that :

$\forall i \in [\![0,m[\![ :$
$$
\begin{equation}
\begin{cases}
\forall k, S_i^0(k) < S_i(p) \\
S_i^1 = {S_i(p)} \\

$$

Let's now rename the subarrays such that : $S_i^j=S_{3i+j}$

$$\forall i \in [\![0,m[\![, \forall (j_1,j_2) \in [\![0,2]\!]^2,(j_1<j_2)\implies (\forall (k_1,k_2),S_{3i+j_1}(k_1) \leq S_{3i+j_2}(k_2)) \tag{E1}$$

And from $(\mathrm{H})$ :
$$\forall (i_1,i_2) \in [\![0,m[\![^2, \forall j \in [\![0,2]\!],(i_1<i_2)\implies (\forall (k_1,k_2),S_{3i_1+j}(k_1) \leq S_{3i_2+j}(k_2)) \tag{E2}$$

So for $k_1<k_2$ with $k_1=3i_1+j_1$ and the same for $k_2$, then :

$$
\begin{equation} \tag{2}
\begin{split}

**Quick Sort :** $T(n) \in O(n\ln(n))$ but might be polynomial because we didn't implement Median of medians

**Selection Sort :** $T(n) = \frac{n^2+9n}{2}$

**Bubble Sort :** $T(n) = \frac{3n^2-3n}{2}$

**Insertion Sort :** $T(n) = 2n^2 -1$

We therefore expect that Merge Sort and Quick Sort be the fastest algorithms and that Selection Sort, Bubble Sort and Insertion Sort be much slower.

````python
import time
import random
import matplotlib.pyplot as plt

def compare(array_sizes,merge,quick,sel,ins,bubble):
    merge_sort_times = [] # create an empty list to hold the merge sort times
    quick_sort_times = [] # create an empty list to hold the quick sort times
    selection_sort_times=[]
    insertion_sort_times=[]
    bubble_sort_times=[]
    # Time the sorting algorithms on arrays of increasing size
    for size in array_sizes:
        arr = [random.randint(-1000, 1000) for i in range(size)] # create a random array of the given size
        start_time = time.time() # record the start time
        if(merge):
            merge_sort(arr) # apply merge sort to the array
        merge_sort_times.append(time.time() - start_time) # record the time taken and add it to the merge sort times list
        start_time = time.time() # record the start time again
        if(quick):
            quicksort(arr,0,len(arr)-1) # apply quick sort to the array
        quick_sort_times.append(time.time() - start_time) # record the time taken and add it to the quick sort times list
        start_time = time.time() # record the start time
        if(sel):
            selSort(arr) # apply selection sort to the array
        selection_sort_times.append(time.time() - start_time) # record the time taken and add it to the selection sort times list
        start_time = time.time() # record the start time
        if(ins):
            insertion_sort(arr) # apply merge sort to the array
        insertion_sort_times.append(time.time() - start_time) # record the time taken and add it to the insertion sort times list
        start_time = time.time() # record the start time
        if(bubble):
            bubbleSort(arr) # apply merge sort to the array
        bubble_sort_times.append(time.time() - start_time) # record the time taken and add it to the bubble sort times list

    # Plot the results on a line chart
    if(merge):
        plt.plot(array_sizes, merge_sort_times, label='Merge Sort') # plot the merge sort times against the array sizes
    if(quick):
        plt.plot(array_sizes, quick_sort_times, label='Quick Sort') # plot the quick sort times against the array sizes
    if(sel):
        plt.plot(array_sizes, selection_sort_times, label='Selection Sort') # plot the selection sort times against the array sizes
    if(ins):
        plt.plot(array_sizes, insertion_sort_times, label='Insertion Sort') # plot the insertion sort times against the array sizes
    if(bubble):
        plt.plot(array_sizes, bubble_sort_times, label='Bubble Sort') # plot the bubble sort times against the array sizes

    plt.xlabel('Array Size') # label the x-axis
    plt.ylabel('Time (seconds)') # label the y-axis
    plt.title('Comparing Sorting Algorithms') # add a title to the chart
    plt.legend() # add a legend to show which line is which
    plt.show() # display the chart
````

````python
# Generate random arrays of increasing size
array_sizes =range(1, 5001, 1000)# define the range of array sizes to use

compare(array_sizes,True,True,True,True,True)
````
